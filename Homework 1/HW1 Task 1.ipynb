```python
df = pd.read_csv('Data_for_students_1.csv', skiprows=1, index_col=0, parse_dates=[0])
f = df['Flow']
t1 = df['Temperature']
t2 = df['Temperature_2']
fig, ax1 = plt.subplots()
ax1.set_xlabel('Time')
ax1.set_ylabel('Flow')
p1 = ax1.plot(f, label='Flow')
ax2 = ax1.twinx()
ax2.set_ylabel('Temperature')
p2 = ax2.plot(t1, label='Temperature', color='tab:orange')
p3 = ax2.plot(t2, label='Temperature_2', color='tab:green')
lb = p1 + p2 + p3
labs = [l.get_label() for l in lb]
ax1.legend(lb, labs, loc="upper left")
plt.gcf().autofmt_xdate()
fig.tight_layout()
plt.show()
```
