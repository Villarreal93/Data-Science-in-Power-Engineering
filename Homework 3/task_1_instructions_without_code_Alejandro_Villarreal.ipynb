```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

**Variable description** (check the burners_dwg.pdf):
* 'FUEL OIL TO BOILER FLOW (T/H)' - sum of all fuel injected to the boiler (from all burners)
* 'BURNER FLAME SENSOR, BURNER 4' - sensor of flame brightness [%]
* 'FUEL OIL FLOW UPSTREAM BOILER (T/H)' - fuel flow to the boiler - **09PF001**
* 'FUEL OIL FLOW DOWNSTREAM BOILER (T/H)' - fuel flow downstream the boiler - **09PF002**
* 'STEAM PRESSURE, RING 1' - **09PP008**
* 'FUEL OIL PRESSURE, RING 1' - **09PP006**
* 'FUEL OIL PRESSURE UPSTREAM OIL FILTER' - **09PP015**
* 'FUEL OIL PRESSURE DOWNSTREAM OIL FILTER' - **09PP017**
* 'I&C AIR PRESSURE' - **09PP082** [kPa]
* 'COOLING AIR PRESSURE' - **09PP084** [kPa]
* 'FUEL OIL PRESSURE DOWNSTREAM THREE-WAY VALVE' - **09PP412** [MPa]
* 'STEAM PRESSURE DOWNSTREAM THREE-WAY VALVE' - **09PP422** [MPa]
* 'FUEL OIL TEMPERATURE (DEG C)' - **09PT005** [°C]
* 'FUEL OIL CONTROL VALVE POSITION, RING 1' - **AR036**
* 'POWER OUTPUT (MW)' - output power of the plant,
* 'STEAM TEMPERATURE' - **09PT004** [°C]

#### Recreate FUEL OIL TO BOILER FLOW measurement using other measurements. Plot the signal you recreated and the actual FUEL OIL TO BOILER FLOW measurement on single plot. Calculate the difference between these two. Plot the absolute difference in time. Remember about axis labels and the legend! Add units if missing.

```python
sub2 = df.copy()
sub2['timestamp'] = pd.to_datetime(sub2['timestamp'])
sub2['rec_signal'] = sub2.apply(lambda x: x['FUEL OIL FLOW UPSTREAM BOILER (T/H)'] - 
                                x['FUEL OIL FLOW DOWNSTREAM BOILER (T/H)'], axis=1)
sub2.plot(x='timestamp', y=['rec_signal', 'FUEL OIL TO BOILER FLOW (T/H)'], 
          ylabel='Fuel oil flow [T/H]', style=['-','-.']);
plt.xlabel(None);
plt.legend(['recreated signal','original signal']);
```

```python
sub2['abs_dif'] = sub2.apply(lambda x: abs(x['FUEL OIL TO BOILER FLOW (T/H)'] - x['rec_signal']), axis=1)
sub2.plot(x='timestamp', y='abs_dif', legend=None, 
         ylabel='Absolute difference between \n recreated signal and original measurement [T/H]');
plt.xlabel(None);
```

#### Find the moment (timestamp) when first attempt to start the burner # 4 was made. In other words - the moment at which the flame was seen by the brightness sensor for the first time.

```python
sub3 = df.copy()
sub3 = sub3[sub3['BURNER FLAME SENSOR, BURNER 4']>0]
print('The first attempt to fire the burner was made at {}'.format(sub3.loc[sub3.index[0],'timestamp']))
```

#### Find the beginnings (timestamps) and ends (timestamp) of all the attempts. Plot them as horizontal lines on the timeseries plot of BURNER FLAME SENSOR measurement for the first 4 attempts. (See figure below)


```python
sub4 = df.copy()
sub4['timestamp'] = pd.to_datetime(sub4['timestamp'])
sub4['active'] = sub4.apply(lambda x: 1 if x['BURNER FLAME SENSOR, BURNER 4'] >= 0 else 0, axis=1)
sub4['active_dif'] = sub4['active'].diff()
sub4 = sub4.set_index('timestamp')
starts = sub4[sub4['active_dif']==1]
ends = sub4[sub4['active_dif']==-1]
margin = dt.timedelta(seconds=55)
plt.figure(figsize=(16,5));
plt.ylabel('Flame sensor brightness [%]');
plt.xlabel('Time');

for n, (start, end) in enumerate (zip(starts.index, ends.index)):
    if n < 4:
        plt.plot(sub4.loc[start-2*margin:end+margin, 'BURNER FLAME SENSOR, BURNER 4'], color='tab:blue', 
                label='BURNER FLAME SENSOR, BURNER 4')

for i, start in enumerate (starts.index): 
    if i < 4:
        plt.axvline(x=start, color='red', label="Attempt Beginning", linestyle='-.')

for j, end in enumerate (ends.index):  
    if j < 4:
        plt.axvline(x=end, color='green', label="Attempt End", linestyle='-.')
    
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='lower left', bbox_to_anchor=(0,1.1));
plt.show()
```

#### For the first 4 attempts, plot flame sensor brightness, fuel pressure and steam pressure downstream the 3-way valve on single plot (see below)

```python
sub5 = df.copy()
sub5['timestamp'] = pd.to_datetime(sub5['timestamp'])
sub5['active'] = sub5.apply(lambda x: 1 if x['BURNER FLAME SENSOR, BURNER 4'] >= 0 else 0, axis=1)
sub5['active_dif'] = sub5['active'].diff()
sub5 = sub5.set_index('timestamp')
starts = sub5[sub5['active_dif']==1]
ends = sub5[sub5['active_dif']==-1]
margin = dt.timedelta(seconds=60)
fig, ax1 = plt.subplots(figsize=(14,4))
ax1.set_ylabel('Flame sensor brightness [%]');
ax2 = ax1.twinx()
ax2.set_ylabel('Pressure [MPa]')
for n, (start, end) in enumerate (zip(starts.index, ends.index)):
    if n < 4:
        p1 = ax1.plot(sub5.loc[start-2*margin:end+margin, 'BURNER FLAME SENSOR, BURNER 4'],  
                 label='Flame sensor' if n==0 else "", color='black')
        p2 = ax2.plot(sub5.loc[start-2*margin:end+margin, 'STEAM PRESSURE DOWNSTREAM THREE-WAY VALVE'],
                 label='STEAM PRESSURE DOWNSTREAM THREE-WAY VALVE', color='tab:blue')
        p3 = ax2.plot(sub5.loc[start-2*margin:end+margin, 'FUEL OIL PRESSURE DOWNSTREAM THREE-WAY VALVE'],
                 label='FUEL OIL PRESSURE DOWNSTREAM THREE-WAY VALVE', color='tab:orange')

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax1.legend(loc='lower left', bbox_to_anchor=(0,1.1));
plt.legend(by_label.values(), by_label.keys(), loc='lower right', bbox_to_anchor=(1,1.1));
plt.show()
```
